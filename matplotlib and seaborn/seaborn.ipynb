```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
```

```python
type(tips)
```

```python
tips.head()
```

```python
sns.color_palette('rocket')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', size='size', palette='bright')
```

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='day', size='size', palette='bright')
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(x='sex', y='tip', data=tips, palette='bright')
```

```python
sns.boxplot(x='day', y='tip', hue='sex', data=tips, palette='bright')
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='bright')
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', data=tips, palette='bright', dodge=True)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', fill=True, cmap='YlGnBu')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```
